## Milestone 2

## Calcule com 95% de confiança a média de dinheiro gasto com saude e educação

```python
#Considerando o teorema do limite central temos que intervalo de confiança = estimador +/- z*SE
Xmed = sample["Medical Care Expenditure"].mean()
SEmed = sample["Medical Care Expenditure"].std()
Interval_min_med = Xmed - (1.96*(SEmed/10))
if Interval_min_med < 0:
    Interval_min_med = 0
else:
    pass
Interval_max_med = Xmed + (1.96*(SEmed/10))

print(f"Com 95% de confiança a média de dinheiro gasto com saúde está entre {Interval_min_med:.2f} e {Interval_max_med:.2f}")

Xedu = sample["Education Expenditure"].mean()
SEedu = sample["Education Expenditure"].std()
Interval_min_edu = Xedu - (1.96*(SEedu/10))
if Interval_min_edu < 0:
    Interval_min_edu = 0
else:
    pass
Interval_max_edu = Xedu + (1.96*(SEedu/10))
print(f'Com 95% de confiança a média de dinheiro gasto com educação está entre {Interval_min_edu:.2f} e {Interval_max_edu:.2f}')
```

## Calcule se há diferença de média de número total de membros da familia pelo sexo do chefe de familia com 95% de confiança

```python
#Realizar teste de hipotese, com alfa = 0.05, sendo H0: Xmale == Xfemale e Ha: Xmale != Xfemale
#Realizar analise de P-valor e verificar se existe evidencia suficiente para rejeitar H0 (se alfa > P-valor, rejeitar H0)
import scipy.stats as st

sample_male = sample[sample["Household Head Sex"]=="Male"]
sample_female = sample[sample["Household Head Sex"]=="Female"]

Xmale = sample_male["Total Number of Family members"].mean()
Xfemale = sample_female["Total Number of Family members"].mean()

SEmale = (sample_male["Total Number of Family members"].std()/(np.sqrt(len(sample_male))))
SEfemale = (sample_female["Total Number of Family members"].std()/(np.sqrt(len(sample_female))))

Zscore = (Xfemale - Xmale)/SEfemale
ProbZ = st.norm.cdf(Zscore)

print(f"""Com 95% de confiança, considerando o P-valor de {ProbZ:.3f}, temos evidencia suficiente para rejeitar a hipotese nula 
H0 que diz que a média de número de pessoas em uma familia é igual para familias lideradas por homens e familias lideradas por
mulheres""")
```

## Calcule se há diferença de média total de renda da familia de acordo com a fonte salarial

```python
#Realizar teste de hipotese, com alfa = 0.05, sendo H0: Xempregado == Xoutro e Ha: Xempregado != Xoutro

import scipy.stats as st

sample_empregado = sample[sample["Main Source of Income"]=="Wage/Salaries"]
sample_outro = sample[sample["Main Source of Income"]!="Wage/Salaries"]

Xempregado = sample_empregado["Total Household Income"].mean()
SEempregado = sample_empregado["Total Household Income"].std()/(np.sqrt(len(sample_empregado)))

SEoutro = sample_outro["Total Household Income"].std()/(np.sqrt(len(sample_outro)))
Xoutro = sample_outro["Total Household Income"].mean()

Zscore = (Xoutro - Xempregado)/SEoutro
ProbZ = st.norm.cdf(Zscore)

print(f"""Com 95% de confiança, considerando o P-valor de {ProbZ:.3f}, temos evidencia suficiente para aceitar a Hipotese Nula
de que a média dos ganhos de um funcionario assalariado é a mesma de um empreendedor""")
```
